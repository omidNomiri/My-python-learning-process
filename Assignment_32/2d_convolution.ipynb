```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
image = cv2.imread("input/spider.png", cv2.IMREAD_GRAYSCALE)
```

```python
edge_detection_kernel = np.array([[-1, -1, -1],
                                  [-1,  8, -1],
                                  [-1, -1, -1]])

result = cv2.filter2D(image, -1, edge_detection_kernel)
result_list = np.hstack((image, result))

plt.imshow(result_list, cmap="gray")
plt.axis(False)
plt.show()
```

```python
sharpening_kernel = np.array([[0, -1,  0],
                              [-1, 5, -1],
                              [0, -1,  0]])

result = cv2.filter2D(image, 0, sharpening_kernel)
result_list = np.hstack((image, result))

plt.imshow(result_list, cmap="gray")
plt.axis(False)
plt.show()
```

```python
emboss_kernel = np.array([[-2, -1,  0],
                          [-1,  1,  1],
                          [0,   1,  2]])

result = cv2.filter2D(image, 0, emboss_kernel)
result_list = np.hstack((image, result))

plt.imshow(result_list, cmap="gray")
plt.axis(False)
plt.show()
```

```python
identity_kernel = np.array([[0,  0,  0],
                            [0,  1,  0],
                            [0,  0,  0]])

result = cv2.filter2D(image, 0, identity_kernel)
result_list = np.hstack((image, result))

plt.imshow(result_list, cmap="gray")
plt.axis(False)
plt.show()
```

```python
black_zone_kernel = np.array([[0,  2,  0],
                              [2, -4,  2],
                              [0,  2,  0]])

result = cv2.filter2D(image, 0, black_zone_kernel)
result_list = np.hstack((image, result))

plt.imshow(result_list, cmap="gray")
plt.axis(False)
plt.show()
```
