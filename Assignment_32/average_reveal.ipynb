```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
hidden_item_image = cv2.imread("input/hidden_item.tif", cv2.IMREAD_GRAYSCALE)
```

```python
def noise_redactor(image, array_size, kernel_size):

    kernel = np.ones((array_size, array_size),
                     dtype=np.uint8) // kernel_size

    result = cv2.filter2D(image, -1, kernel)

    return result
```

```python
avg_kernel_3x3_1 = noise_redactor(hidden_item_image, 3, 1)
avg_kernel_3x3_2 = noise_redactor(hidden_item_image, 3, 0.2)
avg_kernel_3x3_3 = noise_redactor(hidden_item_image, 3, 0.04)

result_list = np.hstack(
    (hidden_item_image, avg_kernel_3x3_1, avg_kernel_3x3_2, avg_kernel_3x3_3))

plt.imshow(result_list, cmap="gray")
plt.axis(False)
plt.show()
```

```python
avg_kernel_5x5_1 = noise_redactor(hidden_item_image, 5, 1)
avg_kernel_5x5_2 = noise_redactor(hidden_item_image, 5, 0.2)
avg_kernel_5x5_3 = noise_redactor(hidden_item_image, 5, 0.04)

result_list = np.hstack(
    (hidden_item_image, avg_kernel_5x5_1, avg_kernel_5x5_2, avg_kernel_5x5_3))

plt.imshow(result_list, cmap="gray")
plt.axis(False)
plt.show()
```
