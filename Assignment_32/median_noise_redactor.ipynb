```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
image_skeleton = cv2.imread(
    "input/noisy_skeleton.png", cv2.IMREAD_GRAYSCALE)

image_board = cv2.imread(
    "input/noisy_board.png", cv2.IMREAD_GRAYSCALE)

image_circle = cv2.imread(
    "input/noisy_image.png", cv2.IMREAD_GRAYSCALE)

image_woman = cv2.imread("input/woman.png", cv2.IMREAD_GRAYSCALE)

image_a = cv2.imread("input/a.png", cv2.IMREAD_GRAYSCALE)

image_man_woman = cv2.imread("input/man_woman.png")
```

```python
def median_noise_redactor(image):
    result = cv2.medianBlur(image, 3)

    result_list = np.hstack((image, result))
    return result_list
```

```python
result = median_noise_redactor(image_skeleton)

plt.imshow(result, cmap="gray")
plt.axis(False)
plt.show()
```

```python
result = median_noise_redactor(image_board)

plt.imshow(result, cmap="gray")
plt.axis(False)
plt.show()
```

```python
result = median_noise_redactor(image_circle)

plt.imshow(result, cmap="gray")
plt.axis(False)
plt.show()
```

```python
result = median_noise_redactor(image_woman)

plt.imshow(result, cmap="gray")
plt.axis(False)
plt.show()
```

```python
result = median_noise_redactor(image_man_woman)
result = cv2.cvtColor(result, cv2.COLOR_BGR2RGB)

plt.imshow(result)
plt.axis(False)
plt.show()
```

```python
result = median_noise_redactor(image_a)

plt.imshow(result, cmap="gray")
plt.axis(False)
plt.show()
```
