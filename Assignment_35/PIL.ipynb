```python
from PIL import Image, ImageOps, ImageDraw, ImageFont
from bidi.algorithm import get_display
import arabic_reshaper
import matplotlib.pyplot as plt
```

```python
monalisa_image = Image.open("input/monalisa.png")
land_image = Image.open("input/land.png")
draw = ImageDraw.Draw(monalisa_image)
```

```python
font = ImageFont.truetype("arial.ttf", 36)
text = ""
reshaped_text = arabic_reshaper.reshape(text)
bidi_text = get_display(reshaped_text)

draw.text((100, 10), bidi_text, (255, 255, 255), font=font)

plt.imshow(monalisa_image)
plt.axis(False)
plt.show()
```

```python
colors = ["red", "green", "blue"]
r, g, b = monalisa_image.split()

for color in colors:
    if color == "red":
        hist = r.histogram()
    elif color == "green":
        hist = g.histogram()
    elif color == "blue":
        hist = b.histogram()
    plt.plot(hist, color=color)

plt.show()
```

```python
equalize_land_image = ImageOps.equalize(land_image)

plt.imshow(equalize_land_image)
plt.axis(False)
plt.show()
```

```python
gray_monalisa_image = ImageOps.grayscale(monalisa_image)

plt.imshow(gray_monalisa_image, "grey")
plt.axis(False)
plt.show()
```
