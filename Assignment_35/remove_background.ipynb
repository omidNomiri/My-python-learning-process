```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
microsoft_logo = cv2.imread("input/microsoft_logo.png")
```

```python
microsoft_logo = cv2.cvtColor(microsoft_logo, cv2.COLOR_BGR2BGRA)

mask = np.all(microsoft_logo[:, :, :3] == [170, 170, 170], axis=-1)
microsoft_logo[mask] = 0

microsoft_logo = cv2.cvtColor(microsoft_logo, cv2.COLOR_BGR2RGB)
plt.imshow(microsoft_logo)
plt.axis(False)
plt.show()
```
