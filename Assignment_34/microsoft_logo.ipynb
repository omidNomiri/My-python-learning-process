```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
width, height = 1000, 1000
R = np.zeros((height, width), dtype=np.uint8)

cv2.rectangle(R, (50, 50), (475, 475), 242, -1)
cv2.rectangle(R, (525, 50), (950, 475), 126, -1)
cv2.rectangle(R, (525, 525), (950, 950), 242, -1)

plt.imshow(R, "gray")
plt.axis(False)
plt.show()
```

```python
G = np.zeros((height, width), dtype=np.uint8)

cv2.rectangle(G, (50, 50), (475, 475), 80, -1)
cv2.rectangle(G, (525, 50), (950, 475), 186, -1)
cv2.rectangle(G, (525, 525), (950, 950), 185, -1)
cv2.rectangle(G, (50, 525), (475, 950), 162, -1)

plt.imshow(G, "gray")
plt.axis(False)
plt.show()
```

```python
B = np.zeros((height, width), dtype=np.uint8)

cv2.rectangle(B, (50, 50), (475, 475), 35, -1)
cv2.rectangle(B, (50, 525), (475, 950), 238, -1)

plt.imshow(B, "gray")
plt.axis(False)
plt.show()
```

```python
logo = cv2.merge([R, G, B])

logo[np.where((logo == [0, 0, 0]).all(axis=-1))] = [170, 170, 170]

plt.imshow(logo, "gray")
plt.axis(False)
plt.show()
```

```python
microsoft_logo = np.ones((2000, 5000, 3), dtype=np.uint8) * 170
microsoft_logo = cv2.cvtColor(microsoft_logo, cv2.COLOR_BGR2RGB)

h, w, _ = logo.shape
microsoft_logo[450:450+h, 500:500+w] = logo

cv2.putText(microsoft_logo, "Microsoft", (1500, 1200), cv2.FONT_HERSHEY_DUPLEX , 23, (255, 255, 255), 50)

plt.imshow(microsoft_logo)
plt.axis(False)
plt.show()
```
