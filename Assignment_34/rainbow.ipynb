```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
width, height = 1000, 1000
R = np.zeros((height, width), dtype=np.uint8)

cv2.ellipse(R, (500, 1000), (500, 500), 0, 180, 360, 255, 50)
cv2.ellipse(R, (500, 1050), (480, 500), 0, 180, 360, 255, 50)
cv2.ellipse(R, (500, 1100), (460, 500), 0, 180, 360, 255, 50)
cv2.ellipse(R, (500, 1300), (360, 500), 0, 180, 360, 127, 50)

plt.imshow(R, "gray")
plt.axis(False)
plt.show()
```

```python
G = np.zeros((height, width), dtype=np.uint8)

cv2.ellipse(G, (500, 1050), (480, 500), 0, 0, 360, 127, 50)
cv2.ellipse(G, (500, 1100), (460, 500), 0, 0, 360, 255, 50)
cv2.ellipse(G, (500, 1150), (440, 500), 0, 0, 360, 255, 50)
cv2.ellipse(G, (500, 1200), (420, 500), 0, 0, 360, 255, 50)

plt.imshow(G, "gray")
plt.axis(False)
plt.show()
```

```python
B = np.zeros((height, width), dtype=np.uint8)

cv2.ellipse(B, (500, 1200), (400, 500), 0, 0, 360, 255, 50)
cv2.ellipse(B, (500, 1250), (390, 500), 0, 0, 360, 255, 50)
cv2.ellipse(B, (500, 1300), (360, 500), 0, 0, 360, 255, 50)

plt.imshow(B, "gray")
plt.axis(False)
plt.show()
```

```python
merged_image = cv2.merge([R, G, B])

plt.imshow(merged_image, "gray")
plt.axis(False)
plt.show()
```
