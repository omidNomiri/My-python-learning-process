```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
rubik_cube = cv2.imread("input/rubik_cube.png")
rubik_cube = cv2.cvtColor(rubik_cube, cv2.COLOR_BGR2RGB)

rubik_cube[np.where((rubik_cube == [0, 255, 255]).all(axis=-1))] = [255, 0, 0]
rubik_cube[np.where((rubik_cube == [255, 255, 0]).all(axis=-1))] = [0, 0, 255]
rubik_cube[np.where((rubik_cube == [255, 0, 255]).all(axis=-1))] = [0, 255, 0]

plt.imshow(rubik_cube)
plt.axis(False)
plt.show()
```
