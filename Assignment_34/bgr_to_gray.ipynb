```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
def bgr2gray(image):
    width, height, _ = image.shape
    gray_image = np.zeros((width, height), dtype=np.uint8)

    for i in range(width):
        for j in range(height):
            gray_value = sum(image[i,j])

            gray_image[i, j] = gray_value // 3

    return gray_image
```

```python
melon_image = cv2.imread("input/melon.png")

gray_melon_image = bgr2gray(melon_image)

plt.imshow(gray_melon_image, "gray")
plt.axis(False)
plt.show()
```
