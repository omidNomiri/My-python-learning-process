```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
melon_image = cv2.imread("input/melon.png")
melon_image = cv2.cvtColor(melon_image, cv2.COLOR_BGR2RGB)

R, G, B = cv2.split(melon_image)

melon_image = cv2.merge([G, R, B])

plt.imshow(melon_image)
plt.axis(False)
plt.show()
```
