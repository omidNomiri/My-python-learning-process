```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
original_spiderman_image = cv2.imread("input/spiderman.png")
```

```python
spiderman_image = original_spiderman_image.copy()

spiderman_hsv = cv2.cvtColor(spiderman_image, cv2.COLOR_BGR2HSV)

H, S, V = cv2.split(spiderman_hsv)

H_new = H.copy().astype(np.float32)

for i in range(spiderman_image.shape[0]):
    for j in range(spiderman_image.shape[1]):
        if H[i , j] < 15 or H[i , j] > 165 :
            if S[i , j] > 150 :
                H_new[i , j] += 30 
        
        elif 60 < H[i , j] < 150 :
            if S[i , j] > 190 and V[i , j] < 165 :
                H_new[i , j] -= 30
                if V[i , j] > 25 :
                    V[i , j] -= 24

result = cv2.merge([H_new.astype(np.uint8), S, V])
```

```python
result = cv2.cvtColor(result, cv2.COLOR_HSV2RGB)

plt.imshow(result)
plt.axis(False)
plt.show()
```
