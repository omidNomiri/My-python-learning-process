```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
original_T_shirt_image = cv2.imread("input\T_shirt.png")
```

```python
T_shirt_image = original_T_shirt_image.copy()
T_shirt_image = cv2.cvtColor(T_shirt_image, cv2.COLOR_BGR2HSV)
H, S, V = cv2.split(T_shirt_image)

H_new = H.copy().astype(np.float32)

for i in range(T_shirt_image.shape[0]):
    for j in range(T_shirt_image.shape[1]):
        if  H[i, j] < 20:
            H_new[i, j] += 100
        elif H[i, j] > 160:
            H_new[i, j] -= 60

H_new[H_new > 180] -= 180
H_new[H_new < 0] += 180

result = cv2.merge([H_new.astype(np.uint8), S, V])
```

```python
result = cv2.cvtColor(result, cv2.COLOR_HSV2RGB)

plt.imshow(result)
plt.axis(False)
plt.show()
```
