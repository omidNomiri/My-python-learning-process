```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
original_superman_image = cv2.imread("input/superman.png")
original_sky_image = cv2.imread("input/sky_city.png")
```

```python
superman_image = original_superman_image.copy()
sky_image = original_sky_image.copy()

superman_hsv = cv2.cvtColor(superman_image, cv2.COLOR_BGR2HSV)
sky_hsv = cv2.cvtColor(sky_image, cv2.COLOR_BGR2HSV)

superman_H, superman_S, superman_V = cv2.split(superman_hsv)
sky_H, sky_S, sky_V = cv2.split(sky_hsv)

for i in range(superman_image.shape[0]):
    for j in range(superman_image.shape[1]):
        if 35 < superman_H[i , j] < 80 and superman_V[i , j] > 85:
            superman_H[i, j] = sky_H[i, j]
            superman_S[i, j] = sky_S[i, j]
            superman_V[i, j] = sky_V[i, j]

result = cv2.merge([superman_H, superman_S, superman_V])
```

```python
result = cv2.cvtColor(result, cv2.COLOR_HSV2RGB)

plt.imshow(result)
plt.axis(False)
plt.show()
```
