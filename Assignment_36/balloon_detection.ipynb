```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
original_balloon_image = cv2.imread("input/balloon.png")
```

```python
balloon_image = original_balloon_image.copy()

HSV_image = cv2.cvtColor(balloon_image, cv2.COLOR_BGR2HSV)
```

```python
lower_hue = np.array([0, 100, 100])
upper_hue = np.array([10, 255, 255])
mask = cv2.inRange(HSV_image, lower_hue, upper_hue)
```

```python
mask_rgb = np.repeat(mask[:, :, np.newaxis], 3, axis=2)

result = cv2.bitwise_and(balloon_image, mask_rgb)

result_rgb = cv2.cvtColor(result, cv2.COLOR_BGR2RGB)

plt.imshow(result_rgb)
plt.axis(False)
plt.show()
```
