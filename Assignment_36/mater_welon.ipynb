```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
original_melon_image = cv2.imread("input/melon.png")
```

```python
melon_image = original_melon_image.copy()

melon_image = cv2.cvtColor(melon_image, cv2.COLOR_BGR2HSV)

H, S, V = cv2.split(melon_image)

H_new = H.copy().astype(np.float32)

for i in range(melon_image.shape[0]):
    for j in range(melon_image.shape[1]):
        if 30 < H[i, j] < 80:
            H_new[i, j] -= 45
        if H_new[i, j] < 0:
            H_new[i, j] += 180
        if H[i, j] < 15 or H[i, j] > 165:
            H_new[i, j] += 60

result = cv2.merge([H_new.astype(np.uint8), S, V])
```

```python
result = cv2.cvtColor(result, cv2.COLOR_HSV2RGB)

plt.imshow(result)
plt.axis(False)
plt.show()
```
